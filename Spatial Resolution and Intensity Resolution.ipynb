```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import io
```

```python
actual_img = cv2.imread(r'Lena.tiff')
```

### Spatial Resolution

```python
def reduce(img):
    return img[::2,::2]
```

```python
img = io.imread('Lena.tiff')

img_256 = reduce(img)

img_128 = reduce(img_256)

img_64 = reduce(img_128)

img_32 = reduce(img_64)

img_16 = reduce(img_32)

img_8 = reduce(img_16)

img_4 = reduce(img_8)

img_2 = reduce(img_4)
```

```python
print(img.shape, img_256.shape, img_16.shape) # shapes 
```

```python
fig=plt.figure(figsize=(32,32))


ax1=fig.add_subplot(9,2,1)
ax1.imshow(img)
ax1.axis('off')

ax1=fig.add_subplot(9,2,2)
ax1.hist(img.ravel(),bins =256)

ax2=fig.add_subplot(9,2,3)
ax2.imshow(img_256)
ax2.axis('off')

ax2=fig.add_subplot(9,2,4)
ax2.hist(img_256.ravel(),bins =256)


ax3=fig.add_subplot(9,2,5)
ax3.imshow(img_128)
ax3.axis('off')


ax3=fig.add_subplot(9,2,6)
ax3.hist(img_128.ravel(),bins =256)

ax4=fig.add_subplot(9,2,7)
ax4.imshow(img_64)
ax4.axis('off')

ax4=fig.add_subplot(9,2,8)
ax4.hist(img_64.ravel(),bins =256)

ax5=fig.add_subplot(9,2,9)
ax5.imshow(img_32)
ax5.axis('off')

ax5=fig.add_subplot(9,2,10)
ax5.hist(img_32.ravel(),bins =256)


ax6=fig.add_subplot(9,2,11)
ax6.imshow(img_16)
ax6.axis('off')

ax6=fig.add_subplot(9,2,12)
ax6.hist(img_16.ravel(),bins =256)

ax7=fig.add_subplot(9,2,13)
ax7.imshow(img_8)
ax7.axis('off')

ax7=fig.add_subplot(9,2,14)
ax7.hist(img_8.ravel(),bins =256)

ax8=fig.add_subplot(9,2,15)
ax8.imshow(img_4)
ax8.axis('off')

ax8=fig.add_subplot(9,2,16)
ax8.hist(img_4.ravel(),bins =256)

ax8=fig.add_subplot(9,2,17)
ax8.imshow(img_2)
ax8.axis('off')

ax8=fig.add_subplot(9,2,18)
ax8.hist(img_2.ravel(),bins =256)

plt.subplots_adjust(wspace=0.01)
plt.show()

```

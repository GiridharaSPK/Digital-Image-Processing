```python
import cv2 as cv
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
```

```python
def reduce(img):
    return img[::2,::2]
```

```python
img = io.imread('ImgDataset/Lena.tiff',0)

img_256 = reduce(img)

img_128 = reduce(img_256)

img_64 = reduce(img_128)

img_32 = reduce(img_64)
```

```python
io.imshow(img)
```

```python
io.imshow(img_256)
```

```python
io.imshow(img_128)
```

```python
io.imshow(img_64)
```

```python
io.imshow(img_32)
```

```python
def toGrey(img):
    return np.divide(np.dot(img,[0.2125, 0.7154, 0.0721]),255)
```

```python
GrImg= toGrey(img)
io.imshow(GrImg)
```

```python
bit_8 = np.divide(GrImg[:,:],2) * 2
io.imshow(bit_8)
```

```python
bit_7 = np.divide(bit_8[:,:],2)*2
io.imshow(bit_7)
```
